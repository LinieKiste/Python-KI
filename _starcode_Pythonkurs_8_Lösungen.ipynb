## handgeschriebene Zahlen erkennen

Ein sehr typisches Beispiel ist die Aufgabe an einen Computer handgeschriebene Zahlen zu erkennen.
Es gibt ein Trainingsset mit ganz vielen handgeschriebenen Zahlen und den Labels dazu (also welche Zahl es seien soll). Diese bekommt der Algorithmus und soll damit lernen.
Mit einem Testset (nur handgeschriebene Zahlen ohne Labels) wollen wir dann prüfen, wie gut der Algorithmus das gelernt hat und ob er unsere handgeschriebenen Zahlen erkennt.
Also los gehts:

Als erstes brauchen wir unsere Daten. Ganz viele handgeschriebene Zahlen und die Labels dazu: <br>
**Anmerkung für Lehrkräfte**: Gerne die Schülerinnen immer wieder daruaf hinweisen sich den Code intensive anzuschauen und zu überlegen, was der Code bedeuten könnte, das evtl auch mit Anderen zu diskutieren.

```python
!pip install -q gradio==2.9b7
import tensorflow as tf
import gradio as gr
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train = x_train / 255.0,
x_test = x_test / 255.0
```

```python
# Zahlen darin sehen zum Beispiel so aus:
from IPython.display import Image, display
display(Image('digits.png', width=600))
```

Als nächstes trainieren wir unser Model. Das heißt der Input sind handgeschriebe Zahlen und Labels und wir möchsten, dass der Algorithmus lernt, was welche Zahl ist. Auf dem nächsten Schaubild sehen wir das nochmal übersichtlicher.

```python
display(Image('training_digit_recogition.png', width=600))
```

```python
# Jetzt trainieren wir unser Model
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128,activation='relu'),
  tf.keras.layers.Dense(10, activation='softmax')
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=6)
```

```python
# GUI um Model auszuprobieren
def classify(input):
    prediction = model.predict(input.reshape(1, 28, 28)).tolist()[0]
    return {str(i): prediction[i] for i in range(10)}

label = gr.outputs.Label(num_top_classes=3)
interface = gr.Interface(fn=classify, inputs="sketchpad", outputs=label)
interface.launch()
```
