## <span style="color: #9C68C8;">Liniendiagramme</span> 
Wie wir in den letzten Wochen gesehen haben, gibt es mehrere Arten, Daten in Python zu visualisieren: durch Text, Tabellen, Diagramme,... Zunächst wollen wir uns nun noch einmal den Liniendiagrammen widmen. Diese Form der Darstellung ist besonders interessant, wenn man den Zusammenhang zwischen zwei Werten kennenlernen möchte. Letzte Woche haben wir die Temperaturen von Planeten in Abhängigkeit vom Abstand zur Sonne betrachtet.

Um einen einfachen Liniengraphen zu erstellen, benötigen wir zwei Listen von Datenpunkten, die die x- und y-Koordinaten repräsentieren. Planst du zum Beispiel, in den Urlaub nach Rom zu fahren, dann möchtest du sicherlich wissen, in welchem Monat die Temperaturen besonders angenehm sind. Dazu kannst du das Ganze wie folgt darstellen:

```python
x = range(1, 13) # Monate von 1 bis 12
y = [8.1, 8.7, 8.7, 11.6, 18.8, 22.8, 25.4, 25.7, 21.4, 17.6, 12.6, 8.9]  # Temperaturen in °C
plt.plot(x, y)
plt.show()
```

🔎 <span style="color: #2E88B6;"> Woher kommt denn jetzt dieses `plt.show()`? Das haben wir doch in der letzten Stunde gar nicht benutzt!  Auch wenn das Ganze in diesem Fall nicht unbedingt notwendig ist, ist es eigentlich sehr wichtig, wenn wir unseren Graphen angezeigt bekommen möchten. In Matplotlib werden nämlich die meisten Einstellungen mit separaten Befehlen angepasst. Damit Python nun weiß, wann der Graph angezeigt werden soll, muss dieser Befehl benutzt werden.  </span>

Das ist jetzt zwar schon ganz cool, aber es fehlen noch ein paar wichtige Dinge, damit dieser Graph nützlich ist... Wenn ihr diesen Graphen einer anderen Person zeigt, würde diese gar nicht wissen, worum es geht. Deshalb müssen wir eine Achsenbeschriftung hinzufügen. Das geht mit Matplotlib ganz einfach! Für die Beschriftung der x-Achse benutzen wir den Befehl `plt.xlabel()` und für die Bescriftung der y-Achse den Befehl `plt.ylabel()`. Zusätzlich können wir noch eine Überschrift mit `plt.title()` einfügen. Das ganze sieht dann wie folgt aus:

```python
plt.plot()
plt.xlabel('x-Achsen Beschriftung')
plt.ylabel('y-Achsen Beschriftung')
plt.title('Überschrift')
```

**<span style="color: #EB6ECB;">Aufgabe 5.1:</span>** *Gebe unserem Monat-Temperaturen-Graphen eine x- und y-Achsen Beschriftung sowie eine Überschrift!*

```python
# Mustelösung

x = range(1, 13) # Monate von 1 bis 12
y = [8.1, 8.7, 8.7, 11.6, 18.8, 22.8, 25.4, 25.7, 21.4, 17.6, 12.6, 8.9]  # Temperaturen in °C
plt.plot(x, y)

plt.xlabel('Monat')
plt.ylabel('Temperatur T[°C]')
plt.title('Temperaturdurchschnitt in Rom')

plt.show()
```

Um das ganze zugänglicher zu machen, können wir den Graphen auch noch ein bisschen vergrößern. Dafür benutzen wir den Befehl `plt.figure()`. Dieser erstellt ein leeres Bild. Das Ganze könnt ihr euch wie eine unbemalte Leinwand oder ein unbedrucktes Blatt Papier vorstellen. Da wir natürlich die Leinwand oder das Papier vor dem Malen oder Drucken brauchen, muss dieser Befehl auch vor allen anderen stehen. Die Größe kann dann mit dem Argument `figsize` angepasst werden. Die Größe wird dabei durch `(B,H)` dargestellt, wobei `B` die Breite und `H` die Höhe ist. Matplotlib füllt dann die gesamte Leinwand bis zum Rand aus! So sieht das dann aus:

```python
plt.figure(figsize=(14,9)) # Dieser Befehl steht immer vor allen Befehlen die die Visualisierung betreffen

plt.plot()
plt.xlabel('x-Achsen Beschriftung')
plt.ylabel('y-Achsen Beschriftung')
plt.title('Überschrift')

plt.show() # Dieser Befehl steht immer nach allen anderen Befehlen die Visualisierung betreffen
```

Das einzige Problem das uns noch bleibt ist, dass zwar das Bild größer ist, aber die Achsenbeschriftung nicht. Dies lässt sich durch das Argument `fontsize` (also Schriftgröße auf englisch) beheben. 

```python
plt.figure(figsize=(14,9))

plt.plot()
plt.xlabel('x-Achsen Beschriftung', fontsize=25)
plt.ylabel('y-Achsen Beschriftung', fontsize=25)
plt.title('Überschrift', fontsize=30)

plt.show()
```

## <span style="color: #9C68C8;">Mehrere Linien in einem Diagramm</span> 

Manchmal möchten wir mehrere Linien in einem Diagramm darstellen, um Daten zu vergleichen. Zum Beispiel könnten wir uns die minimale und die maximale Temperatur in Rom anzeigen lassen. Dazu können wir einfach den Befehl `plt.plot()` mehrfach wiederholen.

```python
x = range(1, 13) # Monate von 1 bis 12
maximale_temperatur          = [13.1, 14.3, 17.1, 19.2, 24.4, 28.7, 31.8, 32.1, 27.1, 22.8, 17.2, 13.3]  # Temperaturen in °C
durchschnittliche_temperatur = [8.1, 8.7, 8.7, 11.6, 18.8, 22.8, 25.4, 25.7, 21.4, 17.6, 12.6, 8.9]  # Temperaturen in °C
minimale_temperatur          = [3, 3, 6.1, 8.2, 13.1, 16.8, 19, 19.3, 15.6, 12.3, 7.9, 4.4]  # Temperaturen in °C

plt.figure(figsize=(14,9))

plt.plot(x, maximale_temperatur)
plt.plot(x, durchschnittliche_temperatur)
plt.plot(x, minimale_temperatur)

plt.xlabel('Monat', fontsize=25)
plt.ylabel('Temperatur T[°C]', fontsize=25)
plt.title('Temperatur in Rom', fontsize=30)

plt.show()
```

Warte... Welche Linie ist jetzt welche? Dafür können wir eine Legende verwenden! Der Befehl `plt.plot()` hat das Argument `label` welches uns erlaubt eine Beschriftung für diese Linie hinzuzufügen. Am Ende kann dann noch mit `plt.legend()` eine Legende eingefügt werden, die uns erklärt wofür die ganzen Beschriftungen stehen:

```python
x = range(1, 13) # Monate von 1 bis 12
maximale_temperatur          = [13.1, 14.3, 17.1, 19.2, 24.4, 28.7, 31.8, 32.1, 27.1, 22.8, 17.2, 13.3]  # Temperaturen in °C
durchschnittliche_temperatur = [8.1, 8.7, 8.7, 11.6, 18.8, 22.8, 25.4, 25.7, 21.4, 17.6, 12.6, 8.9]  # Temperaturen in °C
minimale_temperatur          = [3, 3, 6.1, 8.2, 13.1, 16.8, 19, 19.3, 15.6, 12.3, 7.9, 4.4]  # Temperaturen in °C

plt.figure(figsize=(14,9))

plt.plot(x, maximale_temperatur, label='Maximale Temperatur')
plt.plot(x, durchschnittliche_temperatur, label='Durchschnittliche Temperatur')
plt.plot(x, minimale_temperatur, label='Minimale Temperatur')

plt.legend(fontsize=15)

plt.xlabel('Monat', fontsize=25)
plt.ylabel('Temperatur T[°C]', fontsize=25)
plt.title('Temperatur in Rom', fontsize=30)

plt.show()
```

Wir benutzen hier gleich wieder das Argument `fontsize` um die Größe des Textes anzupassen.

Als letztes wollen wir jetzt noch die Farben und den Linienstil anpassen. Das können wir auch ganz einfach über die Argumente `color` für die Farbe und `linestyle` für den Linienstil regeln. Was steht uns aber zur Auswahl? Dazu schauen wir einfach in der Dokumentation nach:
- die Liste an verfügbaren Farben finder wir unter: https://matplotlib.org/stable/gallery/color/named_colors.html
- die Liste an verfügbaren Linienstilen finder wir unter: https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html.

Wir entscheiden uns für die Durchschnittstemperatur eine schwarze durchgehende Linie (also `color='black', linestyle='-'`), für die minimale Temperatur eine blaue gestrichelte Linie (also `color='blue', linestyle='--'`) und für die maximale Temperatur eine gestrichelte rote Linie (also `color='red', linestyle='--'`) zu verwenden 

```python
x = range(1, 13) # Monate von 1 bis 12
maximale_temperatur          = [13.1, 14.3, 17.1, 19.2, 24.4, 28.7, 31.8, 32.1, 27.1, 22.8, 17.2, 13.3]  # Temperaturen in °C
durchschnittliche_temperatur = [8.1, 8.7, 8.7, 11.6, 18.8, 22.8, 25.4, 25.7, 21.4, 17.6, 12.6, 8.9]  # Temperaturen in °C
minimale_temperatur          = [3, 3, 6.1, 8.2, 13.1, 16.8, 19, 19.3, 15.6, 12.3, 7.9, 4.4]  # Temperaturen in °C

plt.figure(figsize=(14,9))

plt.plot(x, maximale_temperatur, label='Maximale Temperatur', color='red', linestyle='--')
plt.plot(x, durchschnittliche_temperatur, label='Durchschnittliche Temperatur', color='black', linestyle='-')
plt.plot(x, minimale_temperatur, label='Minimale Temperatur', color='blue', linestyle='--')

plt.legend(fontsize=15)

plt.xlabel('Monat', fontsize=25)
plt.ylabel('Temperatur T[°C]', fontsize=25)
plt.title('Temperatur in Rom', fontsize=30)

plt.show()
```

**<span style="color: #EB6ECB;">Aufgabe 5.3:</span>** *Nach der ganzen Arbeit mit dem Plotten wäre ein Entspannungsurlaub direkt am Meer doch etwas schönes. Unten findest du verschiedene Arrays mit der Durchschnittstemperatur an Verschiedenen Orten. Suche dir ein paar Ziele aus und stelle sie in einem Plot dar. ## <span style="color: #9C68C8;">Streudiagramme (Scatterplots)</span> 

Ein weiterer nützlicher Diagrammtyp ist das Streudiagramm (auch Punktwolke genannt)! Dabei werden Punkte anstelle von Linien verwendet, um Daten darzustellen. Diese Art von Diagrammen wird oft benutzt, um sogenannte 'Cluster' zu finden, also Subgruppen von Datenpunkten, die etwas gemeinsam haben. Manchmal möchte man aber auch einfach Zusammenhänge oder Tendenzen erkennen. Wir werden das gleich an einem Beispiel besser verstehen können!

Um so ein Diagramm zu erstellen, können wir den Befehl `plt.scatter()`  benutzen. Das Coole an Matplotlib ist, dass wir fast alles, was wir bisher gelernt haben, genauso anwenden können! 

Schauen wir uns mal ein Beispiel an: Wir plotten die Dichte in Abhängigkeit vom Durchmesser der Planeten in unserem Sonnensystem:

```python
# Reihenfolge: Merkur, Venus, Erde, Mars, Jupiter, Saturn, Uranus, Neptun
Durchmesser = [4879, 12104, 12756, 6792, 142984, 120536, 51118, 49528] # in km
Dichte      = [5429, 5243, 5514, 3934, 1326, 687, 1270, 1638]          # in kg/m^3

plt.figure(figsize=(14,9))
plt.scatter(Durchmesser, Dichte)
plt.title('Zusammenhang zwischen Dichte und Durchmesser', fontsize=30)
plt.xlabel('Durchmesser d[km]', fontsize=25)
plt.ylabel('Dichte ρ[kg/m^3]', fontsize=25)
plt.show()
```

Ohne genau zu wissen welche Planeten hier im Plot genau sind können wir drei eindeutige Gruppen erkenen! Und wisst ihr was cool ist? Diese Gruppen haben eine tatsächliche Bedeutung! Die vier Planeten oben Links sind unsere Gesteinsplaneten (Merkur, Venus, Erde und Mars),unten in der Mitte finden wir die Eisriesen (Uranus und Neptun) und rechts unten finden wir die Gasriesen (Jupiter und Saturn). Wir können mithilfe solcher Punktwolken Klassifizierungen von uns unbekannten Daten finden.

Wie zuvor können nun mehrere Datansätze auf einmal in ein Diagramm geplottet werden. Dazu wiederholen wir einfach wieder den Befehl `plt.scatterplot()`. Wir können auch wieder `label` hinzufügen und mit `plt.legend()` uns eine Legende anzeigen lassen. Die Datensätze lassen sich natürlich auch wieder in unterschiedlichen Farben darstellen, hier allerdings mit dem argument `c` (für *color*). Zusätzlich können wir noch die Größe der Punkte mit `s` (für *size*) anpassen, und das aussehen der Punkte mit `marker` anpassen. Welche Marker zur verfügung stehen kann  in der Dokumenatation nachgeschlagen werden: https://matplotlib.org/stable/api/markers_api.html.

**<span style="color: #EB6ECB;">Aufgabe 5.4:</span>** *In eurem Jahrgang bricht eine große Diskussion aus: Bringt lernen tatsächlich etwas? Ihr beschließt wissenschaftlich an die Sache ranzugehen! Die Matheklausur steht vor der Tür. Ihr wisst, dass insgesamt 100 Punkte erreichbar sein werden. Ihr bittet 50 SchülerInnen die Zeit zu Messen in der sie gelernt haben. Zudem bittet ihr Sie aber auch Gummibärchen nach Farbe zu sortieren und die Zeit zu messen, die Sie damit verbracht haben. Was hat jetzt tatsächlich einen Einfluss? Benutze deine Kentnisse und stelle den Zusammenhang zwischen der erreichter Punktzahl und der Anzahl an Stunden die mit lernen verbracht wurden und dem Zusammenhang zwischen erreichter Punktzahl und der Anzahl an Stunden die mit der Sortierung von Gummibärchen verbracht wurde in einer Punktwolke dar. Benutze Beschriftungen, Farben, etc. nach belieben!*

```python
# Musterlösung 

punkte = [21, 47, 27, 75, 30, 20, 88, 60, 81, 25, 85, 62, 41, 42, 17, 95, 30, 24, 67, 69, 30, 54, 35, 76, 86, 49, 90, 63, 23, 50, 19, 92, 68, 82, 38, 22, 73, 66, 56, 45, 21, 93, 71, 37, 77, 44, 79, 28, 16, 59]
stunden_gelernt = [2.5, 5.1, 3.2, 8.5, 3.5, 1.5, 9.2, 5.5, 8.3, 2.7, 7.7, 5.9, 4.5, 3.3, 1.1, 8.9, 2.5, 1.9, 6.1, 7.4, 2.7, 4.8, 3.8, 6.9, 7.8, 4.2, 9.5, 5.8, 2.3, 4.7, 1.6, 9.0, 6.2, 8.1, 3.9, 2.2, 7.1, 6.3, 5.3, 4.4, 1.8, 9.4, 6.7, 3.7, 7.3, 4.1, 8.0, 3.1, 1.3, 5.7]
stunden_gummibärchen_sortiert = [7.1, 7.6, 1.2, 6.8, 6.6, 1.5, 1.5, 1.8, 1.6, 2.0, 2.2, 7.4, 9.2, 8.6, 7.2, 5.6, 6.8, 9.0, 7.3, 8.5, 2.7, 8.8, 5.2, 0.4, 9.6, 8.9, 0.8, 5.9, 7.1, 5.3, 9.2, 8.1, 9.5, 3.0, 3.5, 9.1, 0.7, 5.1, 0.9, 1.7, 2.4, 8.1, 1.2, 7.0, 0.8, 9.6, 7.6, 5.9, 4.0, 5.0]

plt.figure(figsize=(14,9))

plt.scatter(stunden_gelernt, punkte,               label='Gelernt', c='red', marker='o', s=100)
plt.scatter(stunden_gummibärchen_sortiert, punkte, label='Gummibärchen Sortiert', c='blue', marker='^', s=100)

plt.legend(fontsize=15)

plt.title('Was hat einen Einfluss auf die Testergebnisse?',fontsize=30)
plt.xlabel('Stunden', fontsize=25)
plt.ylabel('Punktzahl',  fontsize=25)
plt.show()
```

### <span style="color: #C26BCA;"> Balkendiagramme </span>

Manchmal möchten wir statt Liniendiagrammen Balkendiagramme verwenden, um Daten zu vergleichen. Diese Form der Darstellung ist besonders geeignet für Datensätze mit wenigen Datenpunkten. Stell dir vor, du hilfst bei deinem Lieblingseisladen im Sommer aus. Um zu wissen welche Eissorten in großen Mengen produziert werden müssen, schaut ihr euch die Verkaufszahlen von diesem und letzten Jahr an:

```python
import numpy as np
import matplotlib.pyplot as plt

geschmäcker = ['Vanille', 'Schokolade', 'Stracciatella', 'Erdbeere', 'Mango']
verkauft_letztes_jahr = [459, 189, 391, 303, 652]
verkauft_dieses_jahr = [679, 575, 677, 146, 801]

plt.figure(figsize=(14,9))

breite = 0.35  # Breite der Balken 
x = np.arange(len(geschmäcker))  # np.arrange funktioniert wie range() nur für Gleitkommazahlen

# Die Argumente sind bar(Position der Balken (Mitte des Balkens), Y-Wert, Breite der Balken, Label)
plt.bar(x - breite/2, verkauft_letztes_jahr, width=breite, label='Vorheriges Jahr')
plt.bar(x + breite/2, verkauft_dieses_jahr,  width=breite, label='Dieses Jahr')

plt.xlabel('Eissorten', fontsize=25)
plt.ylabel('Verkaufszahlen', fontsize=25)
plt.title('Eisverkauf im Vergleich', fontsize=30)

# Ticks bezeichnet die Markierungen (oder Teilstriche) an der jeweiligen Achse
plt.xticks(x, geschmäcker, fontsize=20) # Dieser Befehl ermöglicht anstelle von Zahlen, die Geschmäcker darzustellen. Dabei wird an der Position in 'x' der entsprechende String in 'Geschmäcker' angegeben
plt.xticks(fontsize=20) # Hier vergrößern wir nur die kleinen Marker an der y-Achse

plt.legend(fontsize=20)
plt.show()
```

### <span style="color: #C26BCA;"> Histogramme </span>

Histogramme helfen uns, die Verteilung von Daten zu visualisieren. Man kann sich ein Histogramm folgendermaßen vorstellen: Zunächst ermittelt man die kleinste und die größte Zahl in einem Datensatz und teilt den Bereich dazwischen in eine festgelegte Anzahl gleich großer Intervalle auf (sogenannte `bins`, engl für *Behälter*). Jeder Datenpunkt, der sich in das entsprechende Intervall einordnet, wird in dieses Intervall einsortiert. Am Ende zählt man die Anzahl der Elemente in jedem Intervall. Die ensprechende Häufigkeit stellen wir dann mit Balken dar. Nehmen wir zum Beispiel die Zeiten die 30 Schülerinnen im 100-Meter-Lauf gebraucht haben. Dann können wir mithilfe von Histogrammen schauen wie oft die Zahlen in einem bestimmten Bereich vorkommen. ### <span style="color: #C26BCA;"> Subplots </span>

Manchmal möchten wir mehrere Diagramme nebeneinander anzeigen, um sie leichter vergleichen zu können. Nehmen wir zum Beispiel die Aufgabe von oben mit dem zusammenhang zwischen Noten und erreichte Punktzahl. Das ganze kann auch so dargestellt werden:

```python
punkte = [21, 47, 27, 75, 30, 20, 88, 60, 81, 25, 85, 62, 41, 42, 17, 95, 30, 24, 67, 69, 30, 54, 35, 76, 86, 49, 90, 63, 23, 50, 19, 92, 68, 82, 38, 22, 73, 66, 56, 45, 21, 93, 71, 37, 77, 44, 79, 28, 16, 59]
stunden_gelernt = [2.5, 5.1, 3.2, 8.5, 3.5, 1.5, 9.2, 5.5, 8.3, 2.7, 7.7, 5.9, 4.5, 3.3, 1.1, 8.9, 2.5, 1.9, 6.1, 7.4, 2.7, 4.8, 3.8, 6.9, 7.8, 4.2, 9.5, 5.8, 2.3, 4.7, 1.6, 9.0, 6.2, 8.1, 3.9, 2.2, 7.1, 6.3, 5.3, 4.4, 1.8, 9.4, 6.7, 3.7, 7.3, 4.1, 8.0, 3.1, 1.3, 5.7]
stunden_gummibärchen_sortiert = [7.1, 7.6, 1.2, 6.8, 6.6, 1.5, 1.5, 1.8, 1.6, 2.0, 2.2, 7.4, 9.2, 8.6, 7.2, 5.6, 6.8, 9.0, 7.3, 8.5, 2.7, 8.8, 5.2, 0.4, 9.6, 8.9, 0.8, 5.9, 7.1, 5.3, 9.2, 8.1, 9.5, 3.0, 3.5, 9.1, 0.7, 5.1, 0.9, 1.7, 2.4, 8.1, 1.2, 7.0, 0.8, 9.6, 7.6, 5.9, 4.0, 5.0]

plt.figure(figsize=(24, 10))

plt.suptitle('Was hat einen Einfluss auf die Testergebnisse?',fontsize=40) # Das ist der Titel für das gesamte Bild

plt.subplot(1,2,1) # Die erste Zahl steht für die Nummer an Zeilen, die zweite für die Nummer an Säulen und die dritte für den Index (also der erste Subplot von zwei)
# Alles danach ist Teil des zweiten Bildes
plt.scatter(stunden_gelernt, punkte, c='red', marker='o', s=100)
plt.title('Gelernte Stunden', fontsize=30)
plt.xlabel('Stunden', fontsize=25)
plt.ylabel('Punktzahl',  fontsize=25)

plt.subplot(1,2,2)# Die erste Zahl steht wieder für die Nummer an Zeilen, die zweite für die Nummer an Säulen und die dritte für den Index (also der zweite Subplot von zwei)
# Alles danach ist Teil des zweiten Bildes
plt.scatter(stunden_gummibärchen_sortiert, punkte, c='blue', marker='^', s=100)
plt.title('Gummibärchen zählen', fontsize=30)
plt.xlabel('Stunden', fontsize=25)
plt.ylabel('Punktzahl',  fontsize=25)

plt.show()
```

### <span style="color: #C26BCA;"> 3D-Diagramme </span>


Mit Matplotlib können wir auch 3D-Diagramme erstellen. Zum Beispiel können wir die Positionen von Sternen im Raum in einem 3D-Scatterplot darstellen:

```python
fig = plt.figure(figsize=(20,15))
ax = fig.add_subplot(111, projection='3d') # Hier wird ein sSubplot zu unserer Leinwand hinzugefügt (mit 3D-Eigenschaft)

# Fiktive x, y, z Koordinaten:
x = [-223.43468702026283, 411.85423892196525, 76.3625621905669, -67.52650045659955, 143.17070985823548, 299.63365883739414, 74.28587520084541, -176.6044759622966, 215.6051876381342, 307.5191183594788]
y = [111.6651240679122, -129.51301335847745, 60.05365860371292, 448.7188116565761, 27.850469123861146, 128.97820788277988, 372.2674416615129, -211.42007002354435, -346.2706369955054, -391.29864059635537]
z = [134.76711724466975, 363.71929692578715, -467.1105841892837, 296.5919077273662, -490.8535401359495, 491.9329465833265, 156.13598902287734, 421.27490385048674, -366.011286158778, -441.972246740861]

# Alle weiteren Befehle werden jetzt ausgeführt, indem unser Subplot 'ax' verändert wird
ax.scatter(x, y, z, c='orange', s=100)
ax.scatter(0, 0, 0, c='green', s=1000) # Erde an Position (0, 0, 0)
ax.set_title('Positionen von Sternen im Raum', fontsize=30) # Im Grunde alles äquivalent wie zuvor, nur unter anderem Namen
ax.set_xlabel('X-Koordinate (in Lichtjahren)', fontsize=25)
ax.set_ylabel('Y-Koordinate (in Lichtjahren)', fontsize=25)
ax.set_zlabel('Z-Koordinate (in Lichtjahren)', fontsize=25)

plt.show()
```
